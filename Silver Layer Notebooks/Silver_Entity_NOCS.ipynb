```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```


### Data Reading

```python
#We have 2 types of data in our bronze container - 3 parquet files and 1 csv file
```

#### Reading NOCS data

```python
df = spark.read.format("csv")\
    .option("header", True)\
    .option("inferSchema", True)\
    .load("abfss://bronze@adlsolympicsproject.dfs.core.windows.net/nocs")

#Here we have built our external locations, so it will be easy to access the data from this
```

```python
display(df)
```

```python
#We dont need both country and country long

df = df.drop("country")
display(df)

```

```python
#To store only first word of the tags - we are using split and taking only 1st index of that

df = df.withColumn("tag", split(col("tag"), '-')[0])
```

```python
display(df)
```

#### Writing NOCS data to delta lke

```python
df.write.format("delta")\
    .mode("append")\
    .option("path", "abfss://silver@adlsolympicsproject.dfs.core.windows.net/nocs")\
    .saveAsTable("olympics.silver.nocs")
```
