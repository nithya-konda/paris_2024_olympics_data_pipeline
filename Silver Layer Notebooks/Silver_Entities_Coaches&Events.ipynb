
## Dynamic Data Reading

#### Parameters

```python
dbutils.widgets.text("source_container", ""),
dbutils.widgets.text("sink_container", ""),
dbutils.widgets.text("folder", "")

```

#### Fetching data

```python
source_container = dbutils.widgets.get("source_container")
sink_container = dbutils.widgets.get("sink_container")
folder = dbutils.widgets.get("folder")
```

### Parametrizing Code

```python
df = spark.read.format("parquet")\
                 .load(f"abfss://{source_container}@adlsolympicsproject.dfs.core.windows.net/{folder}")
```

```python
display(df)
```

```python
df.write.format("delta")\
        .mode("append")\
       .option("path" ,f"abfss://{sink_container}@adlsolympicsproject.dfs.core.windows.net/{folder}")\
        .saveAsTable(f"olympics.{sink_container}.{folder}")
```

#### How we achieved the Dynamic Data Reading here

First we created a A list in JSON format data with key values of source_container, sink_container and folder names in a notebook

We created a Job to run that notebook, fetch those values and be given to the current notebook "Silver_Entities_Coaches&Events" as an input, Silver Entities notebook gets these parameters and run in that job itself, so data will be written to our datalake inour expected folders automatically

We are running the notebook in the loop format as that array fetched from Array_input notebook with all the files and folders name

This way we can just add the newly created files in Array_Input notebook and they can be transformed read and written to silver container easily

Note: Here we are doing any transformation. This is specific to only files which doesnt need any transformation. Just reading from bronze and writing to silver
